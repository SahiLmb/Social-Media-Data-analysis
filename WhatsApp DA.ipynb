```python
import re #regular expressions
import pandas as pd
```

## Opening file using file handling

```python
f = open('WhatsApp Chat with 23-24 BE A Comp Div A 23-24.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\u202f[APM]{2}\s-\s'


# for 24hr format pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s-\s'
```

## Process to convert data from string to dataframe

```python
messages = re.split(pattern, data)[1:] #[1:] means one onwards
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_data': dates})

# convert message_date type to datetime format
df['message_data'] = pd.to_datetime(df['message_data'], format='%m/%d/%y, %I:%M\u202f%p - ')

df.rename(columns={'message_data': 'date'}, inplace=True)
                   
df.head()
```

```python
df.shape
```

*total 3342 messages with 2 columns*

```python
# seperate users and messages

users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
            users.append('group_notification')
            messages.append(entry[0])
            
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

## heatmap

```python
df.head()
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```

```python
df.sample(5)
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name', columns='period',values='message',aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```
